```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = 0.5
q = np.arange(1e-10, 1, 1e-4)
D_p_q = p * np.log2(p/q) + (1-p) * np.log2((1-p)/(1-q))
plt.plot(q, D_p_q)
plt.ylabel('D(p||q)')
plt.xlabel('q')
plt.grid()
```

```python
N_range = np.arange(5,1001, 5)
Dpq_vec = np.zeros(len(N_range))
q_vec =  np.zeros(len(N_range))
N = 5
for iteration, N in enumerate(N_range):
  realizations = np.random.choice(2, N)
  q = np.sum(realizations == 1)/N
  D_p_q = p * np.log2(p/q) + (1-p) * np.log2((1-p)/(1-q))
  q_vec[iteration] = q
  Dpq_vec[iteration] = D_p_q

plt.figure()
plt.plot(N_range, Dpq_vec)
plt.ylabel('D(p||q_N)')
plt.xlabel('N')
plt.grid()

plt.figure()
plt.scatter(q_vec, Dpq_vec)
plt.ylabel('D(p||q)')
plt.xlabel('q')
plt.grid()
```
