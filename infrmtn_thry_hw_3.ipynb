```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# (a)
alpha = 0.1
p = np.arange(0,1,0.001)
I_X_Y = (1 - alpha) * (1 - p) * np.log2(1/(1 - p)) +\
        alpha * (1 - p) * np.log2(alpha/(alpha * (1 - p) + p)) +\
        p * np.log2(1 / (alpha * (1 - p) + p))

plt.plot(p, I_X_Y)
plt.xlabel('p')
plt.ylabel('I(X;Y)')
plt.title(f'alpha: {alpha}')
plt.grid()
```

```python
# (b)
p = 0.5
alpha = np.arange(0.001,1,0.001)
I_X_Y = (1 - alpha) * (1 - p) * np.log2(1/(1 - p)) +\
        alpha * (1 - p) * np.log2(alpha/(alpha * (1 - p) + p)) +\
        p * np.log2(1 / (alpha * (1 - p) + p))

plt.plot(alpha, I_X_Y)
plt.xlabel('$\\alpha$')
plt.ylabel('I(X;Y)')
plt.title(f'p: {p}')
plt.grid()
```

```python
# (c)
alpha_vec = np.arange(0.001,1,0.001)
I_X_Y_max = np.zeros(len(alpha_vec))
p = np.arange(0.001,1,0.001)
for idx, alpha in enumerate(alpha_vec):
  I_X_Y = (1 - alpha) * (1 - p) * np.log2(1/(1 - p)) +\
          alpha * (1 - p) * np.log2(alpha/(alpha * (1 - p) + p)) +\
          p * np.log2(1 / (alpha * (1 - p) + p))
  I_X_Y_max[idx] = max(I_X_Y)

plt.plot(alpha_vec, I_X_Y_max)

plt.xlabel('$\\alpha$')
plt.ylabel('max I(X;Y)')
plt.grid()
```
